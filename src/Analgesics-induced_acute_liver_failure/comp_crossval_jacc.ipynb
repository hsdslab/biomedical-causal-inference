# Compare Cross-Validation with Jaccard

Calculates Jaccard similarities to compare validation folds.

```python
import json
import numpy as np
import os
dataset = "Analgesics-induced_acute_liver_failure"
dirpath = os.path.join("/home/hsdslab/Documents/Csabi/Pharma_crossval/DeepCausalPV-master-main/dat", dataset, "proc")
```

```python
with open(os.path.join(dirpath, "jaccards.json"), "r") as f:
    res = json.load(f)
```

```python
def get_best(model1, model2):
    model1_better = {
    "prr": 0,
    "ror": 0,
    "ebgm": 0,
    "all": 0
    }
    
    for key in model1_better.keys():
        model1_better[key] = np.sum([0.05 for k in range(20) if res[model1][key][k] > res[model2][key][k]])
    
    print(f"{model1} is better than {model2} in the following percentage of the time:")
    for metric in model1_better.keys():
        print(f"{metric}: {round(model1_better[metric],2)}")
        
    return None
```

```python
get_best("xgb", "albert_temp")
```

```python
get_best("xgb_cal", "albert_temp")
```
