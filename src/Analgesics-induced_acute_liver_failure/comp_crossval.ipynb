# Compare Cross-Validation

Visualizes and compares cross-validation performance between models.

```python
import json
import numpy as np
import os
dataset = "Analgesics-induced_acute_liver_failure"
dirpath = os.path.join("/home/hsdslab/Documents/Csabi/Pharma_crossval/DeepCausalPV-master-main/dat", dataset, "proc")
```

```python
with open(os.path.join(dirpath, "all_results.json"), "r") as f:
    res = json.load(f)
```

```python
def get_best(model1, model2):
    model1_better = {
    "auc": 0,
    "precision": 0,
    "recall": 0,
    "f1": 0,
    "accuracy": 0,
    "ece": 0
}
    
    for key in res.keys():
        for i, metric in enumerate(["auc", "precision", "recall", "f1", "accuracy"]):
            if res[key][model1][i] > res[key][model2][i]:
                model1_better[metric] += 0.05
        if res[key][model1][5] < res[key][model2][5]:
            model1_better["ece"] += 0.05
    
    print(f"{model1} is better than {model2} in the following percentage of the time:")
    for metric in model1_better.keys():
        print(f"{metric}: {round(model1_better[metric],2)}")
        
    return None
```

```python
get_best("xgb", "albert_temp")
```

```python
get_best("xgb_cal", "albert_temp")
```
